#### Question-2

Write a function `count_vowels(text)` that takes a string as input and returns the number of vowels (a, e, i, o, u - case-insensitive) present in the string.

Example:
```python
text = "Hello World"
vowel_count = count_vowels(text)
print(vowel_count)
# Expected output: 3
```


````python
# Write your code here
def count_vowels(text):
    vowels = "aeiouAEIOU" 
    count = 0
    for i in text:
        if i in vowels:
            count += 1
    return count

text = input("enter the string")
text1 = list(text)
count = count_vowels(text1)
print("Number of vowels:", count)
````

Example:
```python
print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
# Expected output:
# True
# False
```

````python
# Write your code here
def is_anagram(str1,str2):

    l1 = list(str1)
    l2 = list(str2)

    s1 = set(str1)
    s2 = set(str2)
 
    s_union  = set(l1+l2)
    s_inter  = s1.intersection(s2)

    z = s_union.difference(s_inter)


    if not z:
        print("Strings are anagram")
    else :
        print("Strings are not anagram")
        
str1 = input(("Enter the 1st string"))
str2 = input(("Enter the 2nd string"))
result_anagram = is_anagram(str1,str2)

````

#### Question-7

Write a function `rotate_list(lst, k)` that rotates a given list lst to the right by `k` positions in-place. Assume `k` is a non-negative integer and can be greater than the length of the list.

Example:

```python
my_list = [1, 2, 3, 4, 5]
rotate_list(my_list, 2)
print(my_list)
# Expected output: [4, 5, 1, 2, 3]
```

````python
# Write your code here
def rotate_list(lst, k):
    if not lst:
        return
    
    k = k % len(lst)     
    lst[:] = lst[-k:] + lst[:-k]

my_list = [1, 2, 3, 4, 5]
rotate_list(my_list, 2)
print(my_list)


````

#### Question-9

Write a Python program that prints the first `n` rows of Pascal's triangle. `n` is a positive integer and an user input. To know about Pascal's triangle see the [wikipedia article](https://en.wikipedia.org/wiki/Pascal%27s_triangle): 

Example:

```python
n = 5
# Expected output:
#     1
#    1 1
#   1 2 1
#  1 3 3 1
# 1 4 6 4 1
```

````python
# Write your code here
def pascal_triangle(n):
    triangle = []
    for i in range(n):
        row = [1] * (i + 1)
        for j in range(1, i):
            row[j] = triangle[i-1][j-1] + triangle[i-1][j]
        triangle.append(row)
    return triangle

# Take input from user
N = int(input("Enter the number of rows (N): "))

triangle = pascal_triangle(N)

# Print centered triangle
for i, row in enumerate(triangle):
    print(" " * (N - i), end="")  # add leading spaces
    print(" ".join(str(x) for x in row))

````

Write a function `word_counter(s)` which will take a string `s` as input and returns a dictionary with count of words (case insensitive). Remove all the punctuations and special symbols present in the string.  

Example:
```python
s = "This city is a beautiful city. This is where I live"
d = word_counter(s)

# Expected output (output can be in any order): {"this": 2, "city": 2, "is": 2, "a": 1, "beautiful": 1, "where": 1, "i": 1, "live": 1}
```

````python
# Write your code here
def word_cunter(S):
    S1 = S.lower()
    cleaned_string = "".join(char for char in S1 if char.isalnum() or char.isspace())
    print(cleaned_string)
    for i in cleaned_string.split():
        print(f"{i}: {cleaned_string.split().count(i)}")

    
s = "This city is a beautiful city. This is where I live"
word_cunter(s)


````
